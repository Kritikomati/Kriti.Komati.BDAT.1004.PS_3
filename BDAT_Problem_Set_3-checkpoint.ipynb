```python
#Step 1. Import the necessary libraries
import pandas as pd
```

```python
#Step 2. Import the dataset from this address.
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
```

```python
#Step 3. Assign it to a variable called users
users = pd.read_csv(url,sep = '|')
users.head()
```

```python
#Step 4. Discover what is the mean age per occupation
# Calculate and display the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("Mean Age per Occupation:\n", mean_age_per_occupation)
```

```python
#Step 5. Discover the Male ratio per occupation and sort it from the most to the least
# Calculate male ratio per occupation and sort in descending order
male_ratio_per_occupation = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').mean()).sort_values(ascending=False)
print("\nMale Ratio per Occupation (Sorted):\n", male_ratio_per_occupation)
```

```python
#Step 6. For each occupation, calculate the minimum and maximum ages
# Calculate and display minimum and maximum ages per occupation
min_max_ages_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nMinimum and Maximum Ages per Occupation:\n", min_max_ages_per_occupation)
```

```python
#Step 7. For each combination of occupation and sex, calculate the mean age
# Calculate and display mean age for each occupation and sex combination
mean_age_per_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()
print("\nMean Age for Each Occupation and Sex Combination:\n", mean_age_per_occupation_sex)
```

```python
#Step 8. For each occupation present the percentage of women and men
# Calculate the total count of each occupation
occupation_counts = users['occupation'].value_counts()

# Calculate the number of men and women per occupation
men_count_per_occupation = users[users['gender'] == 'M']['occupation'].value_counts()
women_count_per_occupation = users[users['gender'] == 'F']['occupation'].value_counts()

# Calculate the percentage of women and men per occupation
percentage_men = (men_count_per_occupation / occupation_counts) * 100
percentage_women = (women_count_per_occupation / occupation_counts) * 100

# Create a DataFrame to display the percentages
gender_percentage_per_occupation = pd.DataFrame({'Percentage Men': percentage_men, 'Percentage Women': percentage_women})

# Display the percentage DataFrame
print("\nPercentage of Women and Men per Occupation:\n", gender_percentage_per_occupation)
```

```python
# Step 1: Import Necessary Libraries
import pandas as pd
import numpy as np
```

```python
# Step 2: Import the Dataset
data = pd.read_csv('wind.txt', delim_whitespace=True)
# Print the first few rows of the dataset
print("Imported Dataset:\n", data.head())
```

```python
# Step 3: Replace the First 3 Columns by a Datetime Index
data.rename(columns = {'Yr':'year', 'Mo':'month', 'Dy':'day'},inplace =True)
data['year'] = ['19%i' % i for i in data['year']]
combinedate = pd.to_datetime(data[['year','month','day']])
data.insert(loc = 0, column = 'date', value = combinedate)
data = data.drop(columns = ['year','month','day'])
data.head()
```

```python
# Step 4: Fix Years to Include 20th and 21st Century
def fix_year(year):
    if year > 1989:
        return year - 100
    else:
        return year

data['date'] = data['date'].apply(lambda x: x.replace(year=fix_year(x.year)))
print("Fixed Year in Dataset:\n", data.head())
```

```python
# Step 5: Set the Right Dates as the Index
data.set_index('date', inplace=True)
print("Dataset with Proper Index:\n", data.head())
```

```python
# Step 6: Calculate Missing Values
missing_values = data.isnull().sum()
print("Missing Values for Each Location:\n", missing_values)
```

```python
# Step 7: Calculate Non-Missing Values
non_missing_count = data.count().sum()
print("Total Non-Missing Values:", non_missing_count)
```

```python
# Step 8: Calculate Mean Windspeeds
mean_windspeed = data.mean().mean()
print("Mean Windspeeds:", mean_windspeed)
```

```python
# Step 9: Calculate Location Statistics
loc_stats = data.aggregate(['min', 'max', 'mean', 'std'])
print("Location Statistics:\n", loc_stats)
```

```python
# Step 10: Calculate Day Statistics
day_stats = data.resample('D').aggregate(['min', 'max', 'mean', 'std'])
print("Day Statistics:\n", day_stats)
```

```python
# Step 11: Average Windspeed in January for Each Location
january_avg = data[data.index.month == 1].mean()
print("Average Windspeed in January for Each Location:\n", january_avg)
```

```python
# Step 12: Downsample to Yearly Frequency
yearly_data = data.resample('Y').mean()
print("Yearly Frequency Data:\n", yearly_data)
```

```python
# Step 13: Downsample to Monthly Frequency
monthly_data = data.resample('M').mean()
print("Monthly Frequency Data:\n", monthly_data)
```

```python
# Step 14: Downsample to Weekly Frequency
weekly_data = data.resample('W').mean()
print("Weekly Frequency Data:\n", weekly_data)
```

```python
# Step 15: Weekly Statistics for First 52 Weeks
weekly_stats = weekly_data[:52].aggregate(['min', 'max', 'mean', 'std'])
print("Weekly Statistics for First 52 Weeks:\n", weekly_stats)
```

Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import pandas as pd
import matplotlib.pyplot as plt

# Copy the file path directly from the file explorer and paste it here
file_path = r'C:\Users\Kriti\Downloads\us-marriages-divorces-1867-2014.csv'

# Load the data from the local CSV file
data = pd.read_csv(file_path)

# Extract relevant columns
years = data['Year']
marriages_per_capita = data['Marriages']
divorces_per_capita = data['Divorces']

# Create the line plot with style elements
plt.figure(figsize=(8, 6))  # Smaller size
plt.plot(years, marriages_per_capita, label='Marriages per Capita', color='dodgerblue', marker='o', linewidth=2)
plt.plot(years, divorces_per_capita, label='Divorces per Capita', color='salmon', marker='s', linewidth=2)

# Adding grid lines and setting grid style
plt.grid(True, linestyle='--', alpha=0.7)

# Adding labels and title with adjusted font sizes
plt.xlabel('Year', fontsize=10)
plt.ylabel('Per Capita', fontsize=10)
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)', fontsize=12)

# Adding legend with adjusted font size
plt.legend(fontsize=8)

# Adding annotations for peak years
max_marriages_year = years[marriages_per_capita.idxmax()]
max_marriages_value = marriages_per_capita.max()
plt.annotate(f'Max Marriages ({max_marriages_value:.1f})', xy=(max_marriages_year, max_marriages_value),
             xytext=(max_marriages_year + 10, max_marriages_value + 1),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=8)

max_divorces_year = years[divorces_per_capita.idxmax()]
max_divorces_value = divorces_per_capita.max()
plt.annotate(f'Max Divorces ({max_divorces_value:.1f})', xy=(max_divorces_year, max_divorces_value),
             xytext=(max_divorces_year - 20, max_divorces_value + 1),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=8)

# Adjusting spacing and fonts
plt.xticks(rotation=45, fontsize=8)
plt.yticks(fontsize=8)

# Show the plot
plt.tight_layout()
plt.show()

```

Question 9
Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
roman = pd.read_csv('roman-emperor-reigns.csv')

# Group data by cause of death and count the occurrences
death = roman.groupby('Cause_of_Death').count()[['Emperor']]
emperor = pd.Series(death['Emperor'])
cause = pd.Series(death.index)

# Define explode values for slices
explode = (0.1, 0, 0, 0, 0, 0, 0, 0)

# Create a new color scheme
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#ffb366', '#8c8c8c']

# Create a new figure and axis for the pie chart
fig, ax = plt.subplots()
ax.pie(emperor, explode=explode, labels=cause, autopct='%1.1f%%', shadow=False, startangle=90, colors=colors)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Add title and labels
plt.title('Cause of Death - Roman Emperors', fontsize=16, fontweight='bold')
plt.xlabel('Cause of Death', fontweight='bold')
plt.ylabel('Number of Emperors', fontweight='bold')

# Show the pie chart
plt.show()
```

Question 10
Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')  # Update the path to your CSV file

# Extract data columns
years = data['Year']
revenue = data['Total Arcade Revenue (billions)']
doctorates = data['Computer Science Doctorates Awarded (US)']

# Set up the figure and plot
plt.figure(figsize=(10, 6))

# Scatter plot with colored dots based on the year
scatter = plt.scatter(revenue, doctorates, c=years, cmap='coolwarm', s=50, alpha=0.7)

# Add a colorbar to indicate years
plt.colorbar(label='Year')

# Add labels and title
plt.xlabel('Total Arcade Revenue (billions)', fontweight='bold')
plt.ylabel('Computer Science Doctorates Awarded (US)', fontweight='bold')
plt.title('Relationship Between Arcade Revenue and CS Doctorates (2000-2009)', fontweight='bold')

# Display grid
plt.grid(True)

# Label the points with years
for i, year in enumerate(years):
    plt.text(revenue[i] + 0.02, doctorates[i], str(year), fontsize=8, color='black')

# Show the plot
plt.show()
```
